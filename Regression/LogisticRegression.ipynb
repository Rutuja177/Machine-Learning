```python
#https://www.geeksforgeeks.org/implementation-of-logistic-regression-from-scratch-using-python/?ref=rp
from sklearn import datasets
from sklearn.datasets import fetch_openml
#split the data set
from sklearn.model_selection import train_test_split
#to create the combination
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
iris = datasets.load_iris()
```

```python
X = iris.data
y = (iris.target !=0) *1
#y = iris.target[:-50] 
print(X)
print(y)
y.shape, X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
class LoggisticRegression():
    
    def __init__(self, learningRate, iterations):
        self.learningRate = learningRate
        self.iterations = iterations
        
    def fit(self, X,y):
        self.sample_rows, self.sample_feature  = X.shape
        self.w = np.zeros(self.sample_feature)
        self.b = 0
        self.X = X
        self.y = y
        
        for i in range(self.iterations):
            y_predicted = 1/(1+np.exp(-(self.X.dot(self.w)+self.b)))
            #calculating gradient descent 
            temp = (y_predicted-self.y.T)
            temp = np.reshape(temp, self.sample_rows)
            w_gradient= np.dot(self.X.T, temp)/self.sample_rows
            b_gradient= np.sum(temp)/self.sample_rows
            
            #updating weights
            self.w = self.w - self.learningRate*w_gradient
            self.b = self.b - self.learningRate*b_gradient
            
            return self
    def predict(self, X):
        sigmoid = 1/(1+np.exp(-(X.dot(self.w)+self.b)))
        y = np.where(sigmoid > 0.5,1,0)
        return y
            
            
            
```

```python
logiReg = LoggisticRegression(learningRate = 0.01, iterations=100 )
logiReg.fit(X_train, y_train)
y_prediction = logiReg.predict(X_test)

def getAccuracy(y_new, y_p):
    return np.sum(y_new==y_p) / len(y_new)

print("Accuracy is:",getAccuracy(y_test,y_prediction))
```
